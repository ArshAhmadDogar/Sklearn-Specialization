```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
cols=[]
for i in range(0,39):
    cols.append("item"+str(i))

df = pd.read_csv("mybooks.txt",sep=",",header=None,names = cols)
df.head()
```

```python
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.0001, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(association_results[0])
```

```python
file = open('patterns.txt','w')
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    file.writelines(str(item[1])+ " : " + items[0] + "," + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")

file.close()
```
