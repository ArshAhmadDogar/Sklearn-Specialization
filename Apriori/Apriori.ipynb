# Apriori

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
filepath = "books.txt"
with open(filepath, 'r') as f:
    for line in f:
        df = pd.concat( [df, pd.DataFrame([tuple(line.strip().split(';'))])], ignore_index=True )
```

```python
df.head()
```

```python
df.shape
```

```python
transactions = []
for i in range(0, 1850):
 transactions.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.001, min_cinfidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs         =[tuple(result[2][0][0])[0] for result in results]
    rhs         =[tuple(result[2][0][1])[0] for result in results]
    supports    =[result[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts        =[result[2][0][3] for result in results]
    return list (zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n = 25, columns = "Lift") # Sorting According to descending Presedence
```
