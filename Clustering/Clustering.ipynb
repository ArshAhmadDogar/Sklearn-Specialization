```python
import numpy as np
import pandas as pd
from sklearn.datasets._samples_generator import make_blobs 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
features,labels = make_blobs(n_samples=500,centers=4,cluster_std=2.00)
plt.scatter(features[:,0],features[:,1])
```

```python
km_clf = KMeans(n_clusters =4)
km_clf.fit(features)
```

```python
print(km_clf.cluster_centers_)
```

```python
km_clf.labels_
```

```python
#plot the centers
plt.scatter(features[:,0],features[:,1],c=km_clf.labels_,cmap="rainbow")

#plot the centers
plt.scatter(km_clf.cluster_centers_[:,0],km_clf.cluster_centers_[:,1],s=100,c='black')
```

```python
plt.scatter(features[:,0],features[:,1],c=km_clf.labels_,cmap="rainbow")
```
